```python
# =============================================================================
# CREACIÓN ESTRUCTURA SILVER - Managed Tables (Free Edition)
# =============================================================================

dbutils.widgets.dropdown("env", "dev", ["dev", "prod"], "Ambiente")

env = dbutils.widgets.get("env")
catalog = f"{env}_fraud"
silver = f"{catalog}.silver"

spark.sql(f"CREATE SCHEMA IF NOT EXISTS {silver}")

print(f"Creando schema silver en {catalog}\n")

# clientes_silver
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {silver}.clientes_silver (
    id_cliente BIGINT,
    nombre STRING,
    edad_real INT,
    rango_edad STRING,
    genero STRING,
    ciudad STRING,
    estado STRING,
    ingreso_anual DOUBLE,
    rango_ingreso STRING,
    deuda_total DOUBLE,
    puntaje_fico INT,
    rango_fico STRING,
    flag_alta_deuda BOOLEAN,
    ingestion_timestamp TIMESTAMP,
    process_date STRING
) USING DELTA
""")
print("clientes_silver → creada")

# tarjetas_silver
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {silver}.tarjetas_silver (
    id_tarjeta BIGINT,
    id_cliente BIGINT,
    tipo_tarjeta STRING,
    marca_tarjeta STRING,
    limite_credito DOUBLE,
    antiguedad_meses INT,
    tarjeta_en_riesgo BOOLEAN,
    ingestion_timestamp TIMESTAMP,
    process_date STRING
) USING DELTA
""")
print("tarjetas_silver → creada")

# transacciones_silver (la más importante)
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {silver}.transacciones_silver (
    id_cliente BIGINT,
    id_tarjeta BIGINT,
    fecha_completa TIMESTAMP,
    monto DOUBLE,
    categoria_mcc STRING,
    dia_semana STRING,
    hora_pico BOOLEAN,
    monto_relativo DOUBLE,
    es_fraude STRING,
    ingestion_timestamp TIMESTAMP,
    process_date STRING
) USING DELTA
""")
print("transacciones_silver → creada")

# mcc_silver
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {silver}.mcc_silver (
    codigo_mcc INT,
    descripcion STRING,
    categoria_grupo STRING,
    ingestion_timestamp TIMESTAMP,
    process_date STRING
) USING DELTA
""")
print("mcc_silver → creada")

print("\n✅ Estructura Silver creada correctamente")
display(spark.sql(f"SHOW TABLES IN {silver}"))
```
