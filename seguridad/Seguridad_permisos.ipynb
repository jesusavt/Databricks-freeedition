```python
# 1. Definimos los nombres exactos de los grupos que ya tienes creados
grupos = {
    "governance": "group_data_governance",
    "engineers": "group_data_engineers",
    "scientists": "group_data_scientists"
}

catalog = "dev_fraud"

print(f"🔐 Configurando Gobernanza para el catálogo: {catalog}")

try:
    # --- DATA GOVERNANCE ---
    spark.sql(f"GRANT ALL PRIVILEGES ON CATALOG `{catalog}` TO `{grupos['governance']}`")
    print(f"✅ Full Control asignado a {grupos['governance']}")

    # --- DATA ENGINEERS ---
    spark.sql(f"GRANT USE CATALOG ON CATALOG `{catalog}` TO `{grupos['engineers']}`")
    for schema in ["bronze", "silver", "gold"]:
        spark.sql(f"GRANT USE SCHEMA ON SCHEMA `{catalog}`.`{schema}` TO `{grupos['engineers']}`")
        spark.sql(f"GRANT SELECT, MODIFY ON SCHEMA `{catalog}`.`{schema}` TO `{grupos['engineers']}`")
    print(f"✅ DML asignado a {grupos['engineers']}")

    # --- DATA SCIENTISTS ---
    spark.sql(f"GRANT USE CATALOG ON CATALOG `{catalog}` TO `{grupos['scientists']}`")
    spark.sql(f"GRANT USE SCHEMA ON SCHEMA `{catalog}`.`gold` TO `{grupos['scientists']}`")
    spark.sql(f"GRANT SELECT ON SCHEMA `{catalog}`.`gold` TO `{grupos['scientists']}`")
    print(f"✅ Lectura Gold asignada a {grupos['scientists']}")

except Exception as e:
    print(f"❌ Error de permisos: {e}")
    print("Nota: Asegúrate de que tu usuario actual sea ADMIN del Catálogo.")
```
